# Explanation

## front_atoms:
front_atoms should contain a label for each atom in the front (black) area of the image. 

The first label in front_atoms (C1) will be used to make the front center circle. 

## back_atoms:
back_atoms should contain a label for each atom in the back (red) area of the image. 

The first label in back_atoms (C2) will be used to make the back center circle. 

## front_angles:
front_angles are the dihedral angles in degrees between each each atom in front_atoms > 0 and back_atoms[1] (H4)

## back_angles:
back_angles are the dihedral angles in degress between each atom in back_atoms > 1 and front_atom[1] (H1).

back_atoms[1] (H4) will always be represented pointing straight up.
